```python
import numpy as np
```

```python
Input_file = open("Day 9 Input")
Input = []
for line in Input_file:
    Input.append(line.strip())
```

```python
Head = np.zeros((1000,1000))
Head[500][500] = 1
Head_x = 500
Head_y = 500
Tail = np.zeros((1000,1000))
Tail[500][500] = 1
Tail_x = 500
Tail_y = 500

for i in Input:          
       
    if i[0] == 'U':
        for j in np.arange(int(i[2:])):
            Head[Head_y - 1][Head_x] = 1
            Head[Head_y][Head_x] = 0
            Head_y = Head_y - 1
            
            Difference_x = Head_x - Tail_x
            Difference_y = Head_y - Tail_y
            
            Move_x = 0
            Move_y = 0
            if np.abs(Difference_x) > 1:
                Move_x = 1
            if np.abs(Difference_y) > 1:
                Move_y = 1
            
            if Move_x == 1:
                if Difference_y == 0:
                    Tail_x = Tail_x + np.sign(Difference_x)
                else:
                    Tail_x = Tail_x + np.sign(Difference_x)
                    Tail_y = Tail_y + np.sign(Difference_y)
                    
            if Move_y == 1:
                if Difference_x == 0:
                    Tail_y = Tail_y + np.sign(Difference_y)
                else:
                    Tail_x = Tail_x + np.sign(Difference_x)
                    Tail_y = Tail_y + np.sign(Difference_y)   
                
            Tail[Tail_y][Tail_x] = 1
                    
    
    if i[0] == 'D':
        for j in np.arange(int(i[2:])):
            Head[Head_y + 1][Head_x] = 1
            Head[Head_y][Head_x] = 0
            Head_y = Head_y + 1
            
            Difference_x = Head_x - Tail_x
            Difference_y = Head_y - Tail_y

            Move_x = 0
            Move_y = 0
            if np.abs(Difference_x) > 1:
                Move_x = 1
            if np.abs(Difference_y) > 1:
                Move_y = 1
            
            if Move_x == 1:
                if Difference_y == 0:
                    Tail_x = Tail_x + np.sign(Difference_x)
                else:
                    Tail_x = Tail_x + np.sign(Difference_x)
                    Tail_y = Tail_y + np.sign(Difference_y)
                    
            if Move_y == 1:
                if Difference_x == 0:
                    Tail_y = Tail_y + np.sign(Difference_y)
                else:
                    Tail_x = Tail_x + np.sign(Difference_x)
                    Tail_y = Tail_y + np.sign(Difference_y)
                
            Tail[Tail_y][Tail_x] = 1
    
    if i[0] == 'L':
        for j in np.arange(int(i[2:])):
            Head[Head_y][Head_x - 1] = 1
            Head[Head_y][Head_x] = 0
            Head_x = Head_x - 1
            
            Difference_x = Head_x - Tail_x
            Difference_y = Head_y - Tail_y
            
            Move_x = 0
            Move_y = 0
            if np.abs(Difference_x) > 1:
                Move_x = 1
            if np.abs(Difference_y) > 1:
                Move_y = 1
            
            if Move_x == 1:
                if Difference_y == 0:
                    Tail_x = Tail_x + np.sign(Difference_x)
                else:
                    Tail_x = Tail_x + np.sign(Difference_x)
                    Tail_y = Tail_y + np.sign(Difference_y)
                    
            if Move_y == 1:
                if Difference_x == 0:
                    Tail_y = Tail_y + np.sign(Difference_y)
                else:
                    Tail_x = Tail_x + np.sign(Difference_x)
                    Tail_y = Tail_y + np.sign(Difference_y) 
                
            Tail[Tail_y][Tail_x] = 1
    
    if i[0] == 'R':
        for j in np.arange(int(i[2:])):
            Head[Head_y][Head_x + 1] = 1
            Head[Head_y][Head_x] = 0
            Head_x = Head_x + 1
            
            Difference_x = Head_x - Tail_x
            Difference_y = Head_y - Tail_y
            
            Move_x = 0
            Move_y = 0
            if np.abs(Difference_x) > 1:
                Move_x = 1
            if np.abs(Difference_y) > 1:
                Move_y = 1
            
            if Move_x == 1:
                if Difference_y == 0:
                    Tail_x = Tail_x + np.sign(Difference_x)
                else:
                    Tail_x = Tail_x + np.sign(Difference_x)
                    Tail_y = Tail_y + np.sign(Difference_y)
                    
            if Move_y == 1:
                if Difference_x == 0:
                    Tail_y = Tail_y + np.sign(Difference_y)
                else:
                    Tail_x = Tail_x + np.sign(Difference_x)
                    Tail_y = Tail_y + np.sign(Difference_y)
                
            Tail[Tail_y][Tail_x] = 1

            
counter_Head = 0
for i in Head:
    for j in i:
        counter_Head = counter_Head + j
print(counter_Head)

counter_Tail = 0
for i in Tail:
    for j in i:
        counter_Tail = counter_Tail + j
print(counter_Tail)
```
